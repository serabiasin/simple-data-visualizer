```python
with open("train.txt") as f:
  lines = f.readlines()

length_data=len(lines)
```

```python
import numpy as np
class_number=[]
labels=['car', 'Van', 'Truck', 'Person','Biker', 'Tram', 'Motorbike', 'Train', 'Bus', 'Bicycle']

for i in range(length_data):
  buffer_loop=lines[i].split(' ')[1:]

  for x in range(len(buffer_loop)):
    # print(buffer_loop[x])

    if buffer_loop[x].find('\n')!=-1:
    
      new_buffer=buffer_loop[x].replace("\n","")
      class_number.append(int(new_buffer[-1]))
    
    else:
      class_number.append(int(buffer_loop[x][-1]))

class_number.sort()
array_class=np.array(class_number)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

labels=('car', 'Van', 'Truck', 'Person','Biker', 'Tram', 'Motorbike', 'Train', 'Bus', 'Bicycle')
plt.xticks(range(len(labels)),labels)
plt.hist(array_class, bins=35)


plt.gca().set(title='KITTI-PASCAL DATASET DISTRIBUTION CLASSES', ylabel='# Objects');
```

```python
number_object=[]

for x in range(len(labels)):
  number_object.append(np.array(np.where(array_class==x)).shape[1])

number_object=np.array(number_object)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.pie(number_object, labels = labels, autopct='%1.1f%%',pctdistance=0.9,
        shadow=True, startangle=90)
plt.show()
```
